```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error
```

```python
# Определение столбцов данных
WaterCut_real = 3
WaterCut_coriolis = 7
DensityDrop = 9
MassFlow_err = 10
```

```python
# Извлечение признаков и целевой переменной из датасета
x = dataset.iloc[row_start:, [WaterCut_real, WaterCut_coriolis, DensityDrop]].values.astype(float)
y = dataset.iloc[row_start:, [MassFlow_err]].values.astype(float)
```

```python
# Разделение на обучающий и тестовый наборы
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Список степеней полинома, которые мы хотим попробовать
degrees = [2, 3, 4, 5]
```

```python
# Словарь для хранения результатов
results = {}
```

```python
# Цикл по степеням полинома
for degree in degrees:
    # Создание полиномиальных признаков
    poly = PolynomialFeatures(degree=degree)
    x_train_poly = poly.fit_transform(x_train)
    x_test_poly = poly.transform(x_test)

    # Обучение модели линейной регрессии на полиномиальных признаках
    regression_model = LinearRegression()
    regression_model.fit(x_train_poly, y_train)

    # Предсказание на тестовом наборе
    y_predicted = regression_model.predict(x_test_poly)

    # Рассчет средней абсолютной ошибки
    mae = mean_absolute_error(y_test, y_predicted)

    # Сохранение результата
    results[degree] = mae
```

```python
# Вывод результатов
for degree, mae in results.items():
    print(f"Степень полинома {degree}: Mean absolute error: {mae}")
```
